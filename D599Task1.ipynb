```python
import pandas as pd
```

```python
file1 = "Employee Turnover Dataset.csv"
df = pd.read_csv(file1)
```

```python
#df.isnull().info()
```

```python
df['NumCompaniesPreviouslyWorked'] = df['NumCompaniesPreviouslyWorked'].fillna(
    df['NumCompaniesPreviouslyWorked'].median()
)

df['AnnualProfessionalDevHrs'] = df['AnnualProfessionalDevHrs'].fillna(
    df['AnnualProfessionalDevHrs'].median()
)

df['TextMessageOptIn'] = df['TextMessageOptIn'].fillna('Unknown')
```

```python
df['CompensationType'].value_counts()
```

```python
df['JobRoleArea'].value_counts()
```

```python
df['JobRoleArea'] = (
    df['JobRoleArea']
    .str.replace('InformationTechnology', 'Information Technology')
    .str.replace('Information_Technology', 'Information Technology')
    .str.replace('HumanResources', 'Human Resources')
    .str.replace('Human_Resources', 'Human Resources')
)
```

```python
df['PaycheckMethod'] = (
    df['PaycheckMethod']
    

    .str.replace('Mailed Check', 'Mail Check')
    .str.replace('Mail_Check', 'Mail Check')
    .str.replace('MailedCheck', 'Mail Check')
    .str.replace('DirectDeposit', 'Direct Deposit')
    .str.replace('Direct_Deposit', 'Direct Deposit')
)
```

```python
df.dtypes
```

```python
df['HourlyRate '] = (
    df['HourlyRate ']
    .replace('[\$,]', '', regex=True)
    .astype(float))
```
