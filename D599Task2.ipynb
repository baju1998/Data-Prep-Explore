```python
var1 = df[['age', 'bmi', 'height', 'children', 'charges', 'score']].max() - df[['age', 'bmi', 'height', 'children', 'charges', 'score']].min()
print("Range")
print(var1)
var2 = df[['age', 'bmi', 'height', 'children', 'charges', 'score']].var()
print("Variance")
print(var2)
df.describe()
```

```python
sex_stat = df['sex'].value_counts().to_frame(name='Count')
sex_stat['Percentage'] = df['sex'].value_counts(normalize=True)*100
sex_stat
```

```python
from scipy.stats import ttest_ind

smokers = df[df['smoker'] == 'yes']['charges']
non_smokers = df[df['smoker'] == 'no']['charges']

t_stat, p_value = ttest_ind(smokers, non_smokers, equal_var=False)

print("T-stat=", t_stat ,"P-Value=", p_value)
```

```python
from scipy.stats import kruskal
df = pd.read_csv("Health Insurance Dataset.csv")

groups = [g['charges'].dropna().values for _, g in df.groupby('region')]

H, p = kruskal(*groups)

print("KruskalWallis H test (charges by region)")
print("H statistic =", H)
print("p-value     =", p)

summary = df.groupby('region')['charges'].agg(['count','median','mean'])
print("\nCharges by region (count, median, mean):")
print(summary)
```
